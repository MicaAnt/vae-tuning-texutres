```python
extracted = []
```

```python
for fn in file_list:  
    music = ds.init_music(fn)
    data_track, indicator, _ = music.prepare_data(num_bar=2)
    dataset = ds.ArrangementDataset(
        data_track, indicator,
        shift_low=0, shift_high=0,
        num_bar=2, contain_chord=True,
    )

    for i in range(len(dataset)):
        _, _, piano_roll, _, chord, _ = dataset[i]
        song = os.path.splitext(os.path.basename(fn))[0]
        out_fn = f"{song}_{i:03d}.npz"
        path = os.path.join(OUT_DIR, out_fn)
        np.savez_compressed(path, piano_roll=piano_roll, chord=chord)
        extracted.append((out_fn, piano_roll, chord))
```

```python
conteudo_001_000 = np.load("features/001_000.npz")

piano_roll_001_000 = conteudo_001_000["piano_roll"]
harmonia_001_000 = conteudo_001_000["chord"]

print("Piano roll (001_000):", piano_roll_001_000.shape)
print("Harmonia (001_000):", harmonia_001_000.shape)
```

```python
fig, axs = plt.subplots(1, 2, figsize=(12, 4))

# Piano roll
axs[0].imshow(piano_roll_001_000.T, aspect='auto', origin='lower')
axs[0].set_title("Piano Roll â 001_000")
axs[0].set_xlabel("Tempo (32)")
axs[0].set_ylabel("Nota MIDI (128)")

# Harmonia
axs[1].imshow(harmonia_001_000.T, aspect='auto', origin='lower')
axs[1].set_title("Harmonia â 001_000")
axs[1].set_xlabel("Tempo (8)")
axs[1].set_ylabel("Classe (36)")

plt.tight_layout()
plt.show()
```
