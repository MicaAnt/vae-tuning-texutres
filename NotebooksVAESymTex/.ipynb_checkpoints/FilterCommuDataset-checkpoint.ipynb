```python
csv_path = "../midiDataTest/commu_meta.csv"
#csv_path = "./midiDataTest/commu_meta.csv"
```

```python
metaDataCOMMU = pd.read_csv(csv_path, index_col=0)
```

```python
metaDataCOMMU.head()
```

```python
reqDataCOMMU =  metaDataCOMMU[(metaDataCOMMU["time_signature"] == "4/4") &
 (metaDataCOMMU['sample_rhythm'] == "standard")]
```

```python
reqDataCOMMU.sample(n=50, random_state = 42)
```

```python
def filtro_chords_por_beat(df):
    return df[
        df.apply(lambda row: len(ast.literal_eval(row["chord_progressions"])[0]) /
                            int(row["time_signature"].split("/")[0]) /
                            row["num_measures"] == 2,
                 axis=1)
    ]
```

```python
vaeCOMMUData = filtro_chords_por_beat(reqDataCOMMU)
```

```python
vaeCOMMUData.head()
```

```python
print("Dados originais:", metaDataCOMMU.shape)
print("Dados filtrados por mtrica e ritmo", reqDataCOMMU.shape)
print("Dados filtrados por acordes por compasso", vaeCOMMUData.shape)
```

```python
print("N of measures originais:", sorted(metaDataCOMMU["num_measures"].unique()))
print("N of measures filtrados por mtrica e ritmo:", sorted(reqDataCOMMU["num_measures"].unique()))
print("N of measueres filtrados por acordes por compasso:", sorted(vaeCOMMUData["num_measures"].unique()))
```

```python
for col in vaeCOMMUData.columns:
    print(f" Coluna: {col}")
    print(vaeCOMMUData[col].value_counts())
    print()
```
